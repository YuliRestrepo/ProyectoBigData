```python
import io
import sys 
import re
PATH=r'C:\Windows\System32\Untitled Folder\Big Data\Proyecto\BotPrediction'
DIR_DATA=PATH+'data/input'
sys.path.append(PATH) if PATH not in list(sys.path) else None
import os
import pandas as pd
from tqdm import tqdm
from utils.text_analysis import TextAnalysis
```

```python
ta= TextAnalysis('es')
setting={'url': True, 'mention': True, 'emoji': True, 'hastag': True, 'stopwords': False, 'relabel': True}
```

```python
raw_twcs = pd.read_csv('what.csv',encoding='utf-8')
cabecera=["text"]
raw_twcs.columns=cabecera
raw_twcs
```

```python
raw_twcs['text']
```

```python
list_sentences = []
for row in tqdm(raw_twcs['text'].to_list()):
    text = ta.proper_encoding(row)
    texto = ta.clean_text(text, **setting)
    texto= re.sub("url", '', texto)
    texto= re.sub("mention", '', texto)
    texto= re.sub("ma ", '', texto)
    texto= re.sub("si ", '', texto)
    texto= re.sub("tecnologia", '', texto)
    list_sentences.append(texto)
```

```python
list_sentences
```

```python
lista=list_sentences

with open("texto.txt", "w") as output:
    for row in lista:
        output.write(str(row) + '\n')
```
